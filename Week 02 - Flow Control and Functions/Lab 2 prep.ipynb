```python
# change to 'latin' to download latin files
lang = 'english'

local_path = os.path.join('texts', lang)
remote_path = remote_base + lang + '/'

if not os.path.exists(local_path):
    os.mkdir(local_path)

for f in files:    
    remote = remote_path + f
    local = os.path.join(local_path, f)

    # download the remote content
    print('downloading', remote)
    res = requests.get(remote)

    # save to local file
    with open(local, 'wb') as fh:
        print(' -> saving', local)
        fh.write(res.content)
```
